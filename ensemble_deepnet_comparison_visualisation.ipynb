```python
import pandas
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
```

```python
evaluations = read_csv('dataset_deepnet_ensemble/bigml/bigml_evaluations.csv',index_col=0)
```

```python
ensemble_train_accuracy= list(evaluations['ensemble_train_accuracy'])
ensemble_train_auc = list(evaluations['ensemble_train_auc'])
ensemble_test_accuracy = list(evaluations['ensemble_test_accuracy'])
ensemble_test_auc = list(evaluations['ensemble_test_auc'])
deepnet_train_accuracy =list(evaluations['deepnet_train_accuracy'])
deepnet_train_auc = list(evaluations['deepnet_train_auc'])
deepnet_test_accuracy = list(evaluations['deepnet_test_accuracy'])
deepnet_test_auc = list(evaluations['deepnet_test_auc'])
nbr_data = list(evaluations['nbr_data'])
```

```python
fig, ax = plt.subplots(figsize=(8, 6), sharey=True)
ax.plot(nbr_data, deepnet_test_accuracy, label="deepnet")
ax.plot(nbr_data, ensemble_test_accuracy, label="ensemble")
ax.legend()
plt.ylabel('Accuracy', fontsize=20)
plt.xlabel('Amount of data', fontsize=20)
plt.title('comparison deepnet/ensemble accuracy' , fontsize=24)
plt.show()
```

## Comparison deepnet/ensemble accuracy

* We see that if the deepnet is more efficient on a small amount of data, it is quickly exceeded by the ensemble.

* We also note that the progression of the accuracy on the ensemble is more regular

```python
fig, ax = plt.subplots(figsize=(8, 6), sharey=True)
ax.plot(nbr_data, deepnet_test_auc, label="deepnet")
ax.plot(nbr_data, ensemble_test_auc, label="ensemble")
ax.legend()
plt.ylabel('AUC', fontsize=20)
plt.xlabel('Amount of data', fontsize=20)
plt.title('comparison deepnet/ensemble AUC', fontsize=24)

plt.show()
```

## Comparison deepnet/ensemble AUC

* We see that if the deepnet is more efficient on a small amount of data, it is quickly exceeded by the ensemble.

* We can also see that the progression of the AUC on the ensemble is more regular.

* On the deepnet we notice that the progression of the AUC is more regular than that of the accuracy.

```python
fig, ax = plt.subplots(figsize=(8,6), sharey=True)
ax.plot(nbr_data, deepnet_train_accuracy, label="deepnet train")
ax.plot(nbr_data, deepnet_test_accuracy, label="deepnet test")
ax.legend()
plt.ylabel('Accuracy', fontsize=20)
plt.xlabel('Amount of data', fontsize=20)
plt.title('comparison deepnet accuracy train/test set', fontsize=24)
plt.show()
```

## Comparison deepnet accuracy train/test set

* When increasing the amount of data the accuracy of the train set quickly passes under that of the test set.

* On the other hand, the train and the test set seem to follow the same variations.

```python
fig, ax = plt.subplots(figsize=(8,6), sharey=True)
ax.plot(nbr_data, deepnet_train_auc, label="deepnet train")
ax.plot(nbr_data, deepnet_test_auc, label="deepnet test")
ax.legend()
plt.ylabel('AUC', fontsize=20)
plt.xlabel('Amount of data', fontsize=20)
plt.title('comparison deepnet AUC train/test set', fontsize=24)
plt.show()
```

## Comparison deepnet AUC train/test set

* When increasing the amount of data the auc of the train drops rapidly.

* It passes under the test set around 60 000 data.

* Both seem to reach a plateau.

```python
fig, ax = plt.subplots(figsize=(8,6), sharey=True)
ax.plot(nbr_data, ensemble_train_accuracy, label="ensemble train")
ax.plot(nbr_data, ensemble_test_accuracy, label="ensemble test")
plt.ylabel('Accuracy', fontsize=20)
plt.xlabel('Amount of data', fontsize=20)
plt.title('comparison ensemble accuracy train/test set', fontsize=24)
ax.legend()

plt.show()
```

## Comparison ensemble accuracy train/test set

* When increasing the amount of data the accuracy of the train drops rapidly.

* It passes under the test set around 70 000 data.

* Both seem to reach a plateau.

```python
fig, ax = plt.subplots(figsize=(8,6), sharey=True)
ax.plot(nbr_data, ensemble_train_auc, label="ensemble train")
ax.plot(nbr_data, ensemble_test_auc, label="ensemble test")
ax.legend()

plt.ylabel('AUC', fontsize=20)
plt.xlabel('Amount of data', fontsize=20)
plt.title('comparison ensemble AUC train/test set', fontsize=24)

plt.show()
```

## Comparison ensemble AUC train/test set

* When you increase the amount of data, the train's AUC drops rapidly.
.
* The decline slowed down considerably around 60 000 data.

* On the set test after a bigger increase at the start, 
it seems to reach a cruising speed increasing little, but regularly.

* The two curves seem tender to join.
