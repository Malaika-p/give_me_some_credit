We import libs we need for this chapter

```python
from matplotlib import pyplot
from pandas import read_csv
import numpy
from pandas.plotting import scatter_matrix
```

We load the dataset

```python
filename = "./cs-training.csv"
data = read_csv(filename, index_col=0)
```

We select numeric column

```python
data_num = data.select_dtypes(exclude=['object'])
print(data_num)
```

## 6.1 Univariate Plots

### 6.1.1 Histograms

```python
fig, ax = pyplot.subplots(4, 3, figsize=(15, 25))
ax_flat = [sub for line in ax for sub in line]
data_num.hist(ax=ax_flat[0:11], bins=25, log=True, bottom=0.1)
ax[0,0].set_xscale('log')
pyplot.show()
```

### 6.1.2 Density Plots

```python
data_num.plot(kind='density', subplots=True, layout=(4,3), sharex=False, figsize=(15, 25))
pyplot.show()
```

### 6.1.3 Box and Whisker Plots

```python
data_num.plot(kind='box', subplots=True, layout=(4,3), sharex=False, sharey=False, figsize=(15, 25))
pyplot.show()
```

## 6.2 Multivariate Plots

### 6.2.1 Correlation Matrix Plot
Correlation gives an indication of how related the changes are between two variables. If two
variables change in the same direction they are positively correlated.

```python
names = list(data_num)
correlations = data.corr()
fig = pyplot.figure(figsize=(25, 25))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,11,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names, rotation=40)
ax.set_yticklabels(names)
pyplot.show()
```

### Example of creating a generic correlation matrix plot.

```python
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
pyplot.show()
```

### 6.2.2 Scatter Plot Matrix
A scatter plot shows the relationship between two variables as dots in two dimensions, one
axis for each attribute.

```python
scatter_matrix(data_num, figsize=(15, 25))
pyplot.show()
```
