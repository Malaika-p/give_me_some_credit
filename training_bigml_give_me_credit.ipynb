# BigML API call

On each configuration:
- we make a split with sklearn
- we export in csv trains and tests set obtained

- We train 6 ensemble with the different files prepare

```python
project_files['bigML_ressource'] ={}
project_files['bigML_ressource']['project']  = 'project/5ded5ae3f80b166cce000bf8'
project_files['bigML_ressource']['sources']  = {}
project_files['bigML_ressource']['sources']['train']  = {}
project_files['bigML_ressource']['sources']['dev']  = {}
project_files['bigML_ressource']['datasets']  = {}
project_files['bigML_ressource']['models'] = {}
project_files['bigML_ressource']['models']['ensembles'] = {}
project_files['bigML_ressource']['models']['deepnets'] = {}
project_files['bigML_ressource']['evaluation']  = {}
```

```python
api = BigML(project=project_files['bigML_ressource']['project'])
```

Train the different ensemble on BigML with training data

```python
for i in range(1,7):
    print('fields_test'+str(i), 'Fields configuration :', project_files['files']['test_fields_files']['fields_test'+str(i)]['fields'])
    source = api.create_source(project_files['files']['test_fields_files']['fields_test'+str(i)]['train_file_name'])
# waiting for the source to be finished. Results will be stored in `source`
    api.ok(source)
    print('Source created')
    project_files['bigML_ressource']['sources']['fields_test'+str(i)] = source['resource']
# step 3: creating a dataset from the previously created `source`
    dataset = api.create_dataset(source)
# waiting for the dataset to be finished
    api.ok(dataset) 
    print('Dataset created')
    project_files['bigML_ressource']['datasets']['fields_test'+str(i)] = dataset['resource']
# step 5: creating an Ensemble
    ensemble = api.create_ensemble(dataset)
    
# waiting for the ensemble to be finished
    api.ok(ensemble)
    print('Model ensemble created')
    project_files['bigML_ressource']['models']['ensembles']['fields_test'+str(i)] = ensemble['resource']
    
```

### We create source for evaluation

```python
for i in range(1,7):
    print('fields_test'+str(i))
    source = api.create_source(project_files['files']['test_fields_files']['fields_test'+str(i)]['dev_file_name'])
# waiting for the source to be finished. Results will be stored in `source`
    api.ok(source)
    print('Source created')
    project_files['bigML_ressource']['sources']['dev']['fields_test'+str(i)] = source['resource']
    
    dataset = api.create_dataset(source)
    api.ok(dataset)
    print('Dataset created')
    project_files['bigML_ressource']['datasets']['dev']['fields_test'+str(i)] = dataset['resource']
```

Make predictions on dev dataset

```python
project_files['files']['predictions'] = {}
project_files['files']['predictions']['ensembles']  = {}
project_files['files']['predictions']['deepnets']  = {}

for i in range(1,7):
    print('fields_test'+str(i))
    
    dataset = api.get_dataset(project_files['bigML_ressource']['datasets']['dev']['fields_test'+str(i)])
    ensemble = api.get_ensemble(project_files['bigML_ressource']['models']['ensembles']['fields_test'+str(i)])

   # step 10: creating a batch prediction
    batch_prediction = api.create_batch_prediction(ensemble, dataset,
                                                   {"all_fields": True,
                                                    "header": True,
                                                    "confidence": True,
                                                    "probabilities": True})
   # waiting for the batch_prediction to be finished
    api.ok(batch_prediction)
    print('Batch prediction created')
    # downloading the results to your computer
    api.download_batch_prediction(batch_prediction, filename=dossier_predictions+str(i)+'ensemble_'+filename_predictions)
    project_files['files']['predictions']['ensembles']['prediction'+str(i)] = {'dossier': dossier_predictions, 'file_name': str(i)+'ensemble_'+filename_predictions}
    
```

Perform an evaluation of the differents sets

```python
project_files['bigML_ressource']['evaluation']  = {}
project_files['bigML_ressource']['evaluation']['ensembles']  = {}
project_files['bigML_ressource']['evaluation']['deepnets']  = {}

for i in range(1,7):
    print('fields_test'+str(i))
    dataset = api.get_dataset(project_files['bigML_ressource']['datasets']['dev']['fields_test'+str(i)])

    ensemble = api.get_ensemble(project_files['bigML_ressource']['models']['ensembles']['fields_test'+str(i)])
    evaluation = api.create_evaluation(ensemble, dataset)
    api.ok(evaluation)
    print('Evaluation created')
    project_files['bigML_ressource']['evaluation']['ensembles']['fields_test'+str(i)] = evaluation['resource']
    
    print('Accuracy :', evaluation['object']['result']['model']['accuracy'])
    print('Average area under roc curve :', evaluation['object']['result']['model']['average_area_under_roc_curve'])
    print('Confusion Matrix :')
    display(DataFrame(evaluation['object']['result']['model']['confusion_matrix']))
```
